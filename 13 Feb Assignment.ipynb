```python
'''Q1- Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

In Python, Exception is the base class for all the built-in and user-defined exceptions. It provides a set of common methods and attributes that all exceptions can use. When creating a custom exception, we need to inherit from the Exception class to take advantage of these methods and attributes.

By inheriting from the Exception class, our custom exception automatically gets access to the common methods and attributes provided by the Exception class, such as the ability to print out a detailed traceback when an exception occurs. Additionally, by using the Exception class as the base class for our custom exception, we ensure that our custom exception is compatible with the existing exception handling mechanisms in Python.

For example, let's say we want to create a custom exception called CustomError that is raised when a certain condition is met in our code. We can define the CustomError class as follows:'''

class CustomError(Exception):
    pass

'''Here, we inherit from the built-in Exception class to create our custom exception. Now, 
we can raise our CustomError exception whenever the condition we defined is met in our code:'''
if some_condition:
    raise CustomError("An error occurred!")
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

'''The LookupError class is a base class for all the exceptions that occur when a specified key or index cannot be found. It is a subclass of the Exception class and a superclass of several other exception classes.

The KeyError and IndexError are two common exceptions that inherit from the LookupError class.

KeyError is raised when a key is not found in a dictionary. For example:'''
# defining a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# accessing a non-existing key
try:
    value = my_dict['d']
except KeyError:
    print("Key not found in the dictionary")

# Output: Key not found in the dictionary
'''IndexError is raised when an index is out of range of a sequence (such as a list, tuple or string). For example:'''

# defining a list
my_list = [1, 2, 3]

# accessing an element at an index that is out of range
try:
    value = my_list[3]
except IndexError:
    print("Index out of range")

# Output: Index out of range
```
